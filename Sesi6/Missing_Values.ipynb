```python
import pandas as pd
```

```python
# Membaca dan menampilkan data

df = pd.read_csv('titanic.csv')
df.head()
```

```python
# menampilakan informasi data

df.info()
```

```python
# menghitung banyaknya missing values pada setiap kolom

df.isnull().sum()
```

```python
# banyaknya missing values dalam persen

df.isnull().sum() / len(df) * 100.0
```

```python
# menampilkan semua baris dengan missing values pada kolom Age

df[df['Age'].isnull() == True]
```

```python
# menghapus semua baris yang memuat missing values

delete_row = df.dropna(axis=0)
delete_row.info()
```

```python
# menghapus semua kolom yang memiliki missing values

delete_col = df.dropna(axis=1)
delete_col.info()
```

```python
# mengisi missing values pada Age dengan nilai 20

df['Age'].fillna(20)
```

```python
# mengisi missing values pada Age dengan rata-rata dari Age
# dapat juga menggunakan median(), min(), max(), dst

df['Age'].fillna(df['Age'].mean())
```

```python
# mengisi nilai dengan modus

df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df = pd.DataFrame([('Foreign Cinema', 'Restaurant', pd.NA),
                   ('Liho Liho', 'Restaurant', 224.0),
                   (pd.NA, 'bar', 80.5),
                   (pd.NA, 'bar', pd.NA),
                   (pd.NA, 'bar', 65.23),
                   ('Blue Barn', pd.NA, 361.98)],
           columns=('name', 'type', 'AvgBill')
                 )
df
```

```python
# mengisi missing values dengan metode backfill

df['name'].fillna(method='bfill')
```

```python
# mengisi missing values dengan metode front fill

df['name'].fillna(method='ffill')
```
